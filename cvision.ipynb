```python
import cv2
import numpy as np

# Create a sample image (500x500 with gradient colors)
image = np.zeros((500, 500, 3), dtype=np.uint8)
for i in range(500):
    for j in range(500):
        image[i, j] = [i % 256, j % 256, (i+j) % 256]

# Save as demo.jpg
cv2.imwrite('demo.jpg', image)
print("demo.jpg created successfully!")
```

```python
image=cv2.imread('demo.jpg',0)
image=cv2.GaussianBlur(image,(5,5),0)
threshold_image,th=cv2.threshold(image,157,255,cv2.THRESH_TOZERO)
cv2.imshow('image',th)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```
